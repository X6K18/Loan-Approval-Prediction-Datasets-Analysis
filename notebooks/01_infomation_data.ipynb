# 5. **Descriptive Statistics**

### **Variable no_of_dependents**

```python
no_of_dependents_counts = df[' no_of_dependents'].value_counts()
no_of_dependents_counts
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=' no_of_dependents', data=df, palette='Set2', hue=' no_of_dependents')
plt.title('Distribution of Number of Dependents')
plt.xlabel('Number of Dependents')
plt.ylabel('Count')
plt.show()
```

```python
dependents_counts = df[' no_of_dependents'].value_counts().reset_index()
dependents_counts.columns = ['no_of_dependents', 'count']

dependents_counts = dependents_counts.sort_values(by='no_of_dependents')

fig = px.pie(
    dependents_counts, 
    values='count', 
    names='no_of_dependents', 
    title='Distribution of Number of Dependents',
    color_discrete_sequence=px.colors.qualitative.Vivid, 
    template='plotly_dark',                             
    hole=0.4                                           
)

fig.update_traces(
    textposition='inside', 
    textinfo='percent+label',
    marker=dict(line=dict(color='#000000', width=2))
)

fig.update_layout(
    title_x=0.5,
    legend_title_text='Number of Dependents',
    font=dict(size=13)
)

fig.show()
```
